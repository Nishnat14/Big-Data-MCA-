```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df=pd.read_csv('/content/company_sales_data.csv')
```

```python
print(df)

x=df['month_number']
y=df['total_profit']
plt.plot(x,y)
plt.title('line plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x=df['month_number']
y1=df['bathingsoap']
y2=df['facecream']
y3=df['facewash']
y4=df['moisturizer']
y5=df['toothpaste']

plt.plot(x, y1, label='soap')
plt.plot(x, y2, label='cream')
plt.plot(x, y3, label='facewash')
plt.plot(x, y4, label='moisturizer')
plt.plot(x, y5, label='toothpaste')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiline Plot ')

plt.show()
```

```python
#cate=[max(df['shampoo']),max(df['facewash']),max(df['bathingsoap']),max(df['shampoo'])]
plt.bar(df['facewash'],df['month_number'])
bar_width = 7
plt.title('Bar Chart Example')
plt.xlabel('soap')
plt.ylabel('months')
plt.show()

```

```python
plt.scatter(df['shampoo'], df['facewash'], c=df['total_profit'], s=df['month_number'], cmap='viridis', alpha=0.7)

plt.colorbar(label='profit')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with 4 Attributes')

plt.show()
```
