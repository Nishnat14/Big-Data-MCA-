```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(2)

# Define the number of samples and features
num_samples = 600
num_features = 3

# Generate random feature values (X)
X = np.random.rand(num_samples, num_features)

# Generate random coefficients for each feature
coefficients = np.random.rand(num_features)

# Generate random noise for the target variable
noise = np.random.randn(num_samples)

# Calculate the target variable (y) based on the features and coefficients
y = np.dot(X, coefficients) + noise


```

```python
feature_index = 0
X_feature = X[:, feature_index].reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X_feature, y, test_size=0.2, random_state=3)

model = LinearRegression()

model.fit(X_train, y_train)

# Making predictions on the testing data
y_pred = model.predict(X_test)
```

```python
# Plotting the regression line
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.xlabel('Feature {}'.format(feature_index+1))
plt.ylabel('Target Variable')
plt.title('Regression Line for Feature {}'.format(feature_index+1))
plt.show()
```
